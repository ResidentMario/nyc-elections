We can capture what we want by reading one of two patterns:

1. A block of delimited candidate names in all caps, followed by a block of delimited vote counts.
2. A mixed block of delimited character names interspersed with their (delimited) vote counts.

Let's work with this a bit...

```python
sample_output = !pdf2txt.py "../data/sample_cer.pdf"
```

Let's glob by block-of-text.

```python
globs = []
globbed_lines = []
for line in sample_output:
    if line != '':
        globbed_lines.append(line.strip())
    else:
        globs.append(globbed_lines)
        globbed_lines = []
# newline can appear in sequence, so remove empty lines.
globs = [glob for glob in globs if glob]
```

```python
len(globs)
```

```python
globs
```

A glob with numbers in it should always contain things that can be cast to an `int` in it (plus minus an apostraphe:  `,`). 

A glob with names is trickier, because, remember, a glob might contain nothing except for a single name. What counts as a name? Just about anything! 

In the case that the numbers and names are seperated into seperate globs, we can detect the numbers glob and then track back one to get the names.

In case they are not, we ask if the glob contains `WRITE-IN`, since the split necessary to do this should only appear amongst write-ins (all ballotted candidates start at the top of list, which is given a fresh page to begin on).

<!-- The trick I used is that names always appear on the ballot in uppercase. If I remove certain intermediate strings from the candidate name-string, and then filter out the `'BOARD OF ELECTIONS'` string, I can get closer. -->

```python
import re
re_nums = re.compile("\d")

globs_f = []

for ind, glob in enumerate(globs):
    if glob[0].replace(",", "").isdigit():
        globs_f.append(glob)
    elif "(WRITE-IN)" in " ".join(glob):
        globs_f.append(glob)
```

```python
globs_f
```

Merge globs of only numbers onto the previous glob.

```python
i = 0
valid_globs_ff = []

while i < len(globs_f):
    if (i < len(globs_f) - 1) and all(v for v in map(lambda v: v.replace(",", "").isdigit(), globs_f[i + 1])):
        vote_tuples = [(g[0], g[1]) for g in zip(globs_f[i], globs_f[i + 1])]
        valid_globs_ff.append(vote_tuples)
        i += 1
    else:
        i += 1
```
